# Pandas Options and Settings

```python
# Load required libraries
import pandas as pd
import numpy as np
```

```python
# create dataframe with random data to use as example
example_df = pd.DataFrame(np.random.rand(100,30))
example_df
```

## Max rows / columns

The default max number of rows displayed is 60 and the the default max number of columns displayed is 20.

```python
print(pd.options.display.max_rows)
print(pd.options.display.max_columns)
```

If you have a dataframe with more rows/columns, the middle rows/columns will be truncated when the dataframe is displayed (see above example).

You can set the max_rows/max_columns option larger if you want to display more rows.

```python
pd.options.display.max_columns = 110
example_df
```

## Display Precision

```python
pd.set_option("precision", 4)
example_df
```

## Reset options to default

```python
# reset all options using a regex
pd.reset_option("^display")
example_df
```

```python
# reset individual options
pd.reset_option("max_rows")
pd.reset_option("min_rows")
pd.reset_option("max_columns")
```
